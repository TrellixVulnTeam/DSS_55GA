## 2. Tomando datos de la red Social Google+, recolecte, procese y vizualice los datos con  las siguientes restricciones: ##
### a) Obtener las tres ultimas actividades de X usuario ###
### b) Analizar el texto de una de estas actividades. (Palabras utilizadas, cantidad de veces, etc) ###

```python
# -*- coding: utf-8 -*-
from __future__ import print_function
from apiclient import discovery
from prettytable import PrettyTable
import re

TMPL = '''
    User: %s
    Date: %s
    Titulo: %s
    Post: %s
'''

def listaPalabrasDicFrec(listaPalabras):
    frecuenciaPalab = [listaPalabras.count(p) for p in listaPalabras]
    return dict(zip(listaPalabras,frecuenciaPalab))

def ordenaDicFrec(dicfrec):
    aux = [(dicfrec[key], key) for key in dicfrec]
    aux.sort()
    aux.reverse()
    return aux

API_KEY = 'AIzaSyBymtJ6VWF_yUmdFV8t3y9GL36LIDC70Ic'
SERVICE = discovery.build('plus', 'v1', developerKey=API_KEY)

items = SERVICE.people().search(query='Mauricio Macri').execute().get('items', [])

id = items[0]['id']

resultado = SERVICE.activities().list(userId= id,collection='public',maxResults='3').execute()
publicaciones = resultado.get('items', [])

for data in publicaciones:
    post = ' '.join(data['title'].strip().split())
    if post:
        print(TMPL % (data['actor']['displayName'],data['published'], post, data['object']['content']))      
        lista = re.sub('<[^<]+?>', '',data['object']['content'])
        diccionario = listaPalabrasDicFrec(lista.split())
        diccOrdenado = ordenaDicFrec(diccionario)
            
        table = PrettyTable(field_names=['Palabras', 'Frecuencia'])
        list(map(lambda x: table.add_row(x), diccOrdenado))
        print(table)   
            
            
# Fuente http://wescpy.blogspot.com.ar/2014/09/simple-google-api-access-from-python.html
```
