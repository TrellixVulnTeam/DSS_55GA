## . Sacando información de Microformatos 1 en la Web. ##
### a) Obtener y listar un subconjunto de noticias relevantes usando hNews. ###
#### https://www.kuechen-atlas.de/news/ ####

```python
# -*- coding: utf-8 -*-
import datetime
import mf2py
import json
import requests
import re
from bs4 import BeautifulSoup
from hcalendar import hCalendar
```

```python
# https://www.kuechen-atlas.de/news/
# El formato de las hnews utilizado es una evolucion del formato h-atom del nuevo estandar de microformatos2
# mas informacion http://microformats.org/wiki/h-entry

MAX_HNEWS = 3

def obtenerHnews(lista_hnews, max_hnews):
    
    TMP = '''
    ------------------------------------------------------------------------------------------------------------
    | * Autor: %s
    | * Tipo: %s
    | * Url: %s
    ------------------------------------------------------------------------------------------------------------
    | Contenido
    ------------------------------------------------------------------------------------------------------------
    %s
    ------------------------------------------------------------------------------------------------------------
    '''
    resultado = []

    for i in range(max_hnews):
        try:
            url = "https://www.kuechen-atlas.de" + lista_hnews[i]
            x = mf2py.Parser(url = url).to_dict(filter_by_type = "h-entry")
        
            autor = x[0]['properties']['author'][0]['properties']['name'][0]
            tipo = x[0]['type'][0]
            contenido = x[0]['properties']['content'][0]['value']
            r = TMP % (autor, tipo, url, contenido)
            resultado.append(r)
        except:
            pass  
    return resultado
```

```python
req = requests.get('https://www.kuechen-atlas.de/news/')
soup = BeautifulSoup(req.text, "lxml")
hnews = []

for enlaces in soup.find_all('a'):
    try:
        if re.match('\/news\/\w+',enlaces['href']):
            hnews.append(enlaces['href'])
    except:
        pass
    
# lista de hnews sin elementos repetidos
lista = list(set(hnews))
lista_hnews = obtenerHnews(lista,MAX_HNEWS)

# listo resultados
for l in lista_hnews:
    print ("%s" % l)
```

### b) Listar calendario de eventos.  ###

```python
# -*- coding: utf-8 -*-
url = "http://tantek.com/"
req = requests.get(url)
soup = BeautifulSoup(req.text, "lxml")

EVENT = '''
-------------------------------------------------------------------------------------------------------------
| * Evento: %s
| * Inicio: %s
| * Fin: %s
| * Lugar: %s
-------------------------------------------------------------------------------------------------------------
'''
hcal = hCalendar(soup)

for cal in hcal:
    for event in cal:
        
        inicio = event.dtstart.strftime("%d/%m/%Y")
        try:
            fin = event.dtend.strftime("%d/%m/%Y")
        except:
            fin = '--/--/--'
             
        print (EVENT % (event.summary,inicio,fin,event.location) )
      
```
