```python
def minimo_seguidores(usuario_tweepy):
    return usuario_tweepy.followers_count >= 10 

#usuario = api.me()
usuario = api.get_user('RTFMCelia')
seguidores = filter(minimo_seguidores, usuario.followers())
seguidores
```

#### Datos necesarios ####
  - usuario.followers = [ User ]
  - cantidad seguidores = user.followers_count
  - cantidad de tweets = user.statuses_count
  - obtener tweets de un usuario = user.timeline()
  - obtener retweets de un Tweet = tweet.retweets()

```python
class Clasificador:
    @staticmethod
    def get_color(cant_seguidores):
        if cant_seguidores < 400:
            return "blue"
        elif cant_seguidores >= 400 and cant_seguidores < 2000:
            return "yellow"
        else:
            return "red"
        
    @staticmethod
    def get_tamanio(cant_tweets):
        if cant_tweets < 500:
            return 10
        elif cant_tweets > 500 and cant_tweets < 2000:
            return 15
        else:
            return 20
        
```

### 2) Teniendo en cuenta los pasos realizados en el punto anterior (Adquisicion, preparacion,estructuracion pre visualizacion y visualizacion). Tome n con n mayor a 10 coordenadas geograficas y grafıquelas en un mapa. ###

#### Graficamos ####
  - http://bokeh.pydata.org/en/0.12.4/docs/user_guide/concepts.html#userguide-concepts
  - https://bokeh.pydata.org/en/latest/docs/gallery/iris.html

```python

colores = []
tamanios = []
followers = []
tweets = []

for seguidor in seguidores:
    colores.append(Clasificador.get_color(seguidor.followers_count)) 
    tamanios.append(Clasificador.get_tamanio(seguidor.statuses_count))
    followers.append(seguidor.followers_count)
    tweets.append(seguidor.statuses_count)

p = figure(title = "Clasificador de tweets")
p.xaxis.axis_label = 'Seguidores'
p.yaxis.axis_label = 'Tweets'

p.circle(
    followers, 
    tweets,
    color=colores, 
    fill_alpha=0.2, 
    size=tamanios)

show(p)
```

```python
todos_los_tweets = []

for seguidor in seguidores:
    try:
        statuses = seguidor.timeline()
        for s in statuses:
            todos_los_tweets.append(s)        
    except:
        print "error"
        

len(todos_los_tweets)
```

```python
todos_los_retweets = []

try:
    for tweet in todos_los_tweets:
        for retweet in tweet.retweets():
            todos_los_retweets.append(retweet)
except:
    print "error"
        

len(todos_los_retweets)
```

```python
json_tweets = map(lambda t: t._json, todos_los_tweets)
json_retweets = map(lambda t: t._json, todos_los_retweets)
print len(json_tweets)
print len(json_retweets)
```

```python
with open("bokeh_server/datos/todos_los_tweets.json", "w") as f:
    json.dump(json_tweets, f, indent=5)

with open("bokeh_server/datos/todos_los_retweets.json", "w") as f:
    json.dump(json_retweets, f, indent=5)
print "ok"
```
